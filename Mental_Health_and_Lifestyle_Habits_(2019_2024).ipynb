## Data loading

### Subtask:
Load the dataset into a Pandas DataFrame.


**Reasoning**:

```python
import pandas as pd

df = pd.read_csv('Mental_Health_Lifestyle_Dataset.csv')
display(df.head())
```

## Data exploration

### Subtask:
Explore the dataset to understand its structure, including the number of rows and columns, data types of each column, and the presence of missing values.


**Reasoning**:
Explore the dataset's structure by examining its shape, data types, missing values, and unique values in each column.



```python
# Determine the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Obtain the data types of each column
print("\nData Types of Columns:\n", df.dtypes)

# Identify the number of missing values in each column
print("\nNumber of Missing Values per Column:\n", df.isnull().sum())

# Explore the unique values in each column (limiting to a few columns for brevity)
for column in ['Country', 'Gender', 'Exercise Level', 'Diet Type', 'Mental Health Condition']:
  print(f"\nUnique values in '{column}':\n{df[column].unique()}")
```

**Reasoning**:
Summarize the findings from the previous exploration of the dataset into a clear and concise format.



```python
print("Dataset Summary:")
print("-" * 20)
print(f"The dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
print("It includes information on various aspects such as:")
print("- Demographics (Country, Age, Gender)")
print("- Lifestyle factors (Exercise Level, Diet Type, Sleep Hours, Screen Time per Day)")
print("- Mental health (Mental Health Condition, Stress Level)")
print("- Well-being (Social Interaction Score, Happiness Score)")
print("\nData Types:")
print("-" * 20)
print(df.dtypes)
print("\nMissing Values:")
print("-" * 20)
print(df.isnull().sum())
print("\nNote: The 'Mental Health Condition' column has a significant number of missing values.")
```

## Data analysis

### Subtask:
Calculate descriptive statistics for the numerical columns and identify the unique values for categorical columns.


**Reasoning**:
Calculate descriptive statistics for numerical columns and identify unique values for categorical columns.



```python
# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Descriptive statistics for numerical columns
print("Descriptive Statistics for Numerical Columns:\n")
print(df[numerical_cols].describe())

# Unique values for categorical columns
print("\nUnique Values for Categorical Columns:\n")
for col in categorical_cols:
  print(f"{col}: {df[col].unique()}")
```

## Data visualization

### Subtask:
Visualize the distribution of numerical variables and the frequency of categorical variables.


**Reasoning**:
Create histograms for numerical variables to visualize their distribution.



```python
import matplotlib.pyplot as plt

numerical_cols = ['Age', 'Sleep Hours', 'Work Hours per Week', 'Screen Time per Day (Hours)', 'Social Interaction Score', 'Happiness Score']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 3, i + 1)
    plt.hist(df[col], bins=20)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

**Reasoning**:
Create box plots for numerical variables to visualize their central tendency, spread, and potential outliers.



```python
import matplotlib.pyplot as plt

numerical_cols = ['Age', 'Sleep Hours', 'Work Hours per Week', 'Screen Time per Day (Hours)', 'Social Interaction Score', 'Happiness Score']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 3, i + 1)
    plt.boxplot(df[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

**Reasoning**:
Create bar charts for categorical variables to visualize the frequency of each unique value.



```python
import matplotlib.pyplot as plt

categorical_cols = ['Country', 'Gender', 'Exercise Level', 'Diet Type', 'Stress Level', 'Mental Health Condition']

plt.figure(figsize=(15, 15))
for i, col in enumerate(categorical_cols):
    plt.subplot(3, 2, i + 1)
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Frequency of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
